````python
### Librerias
import os
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

from google.colab import drive
````

````python
# Crear matriz de confusion con los datos presentados para el ejercicio
cf_matrix=np.array([[87, 16],
                    [17, 59]])

cf_matrix
````

````python
# Calcular la precision (Accuracy Score)
accuracy=round(np.trace(cf_matrix)/cf_matrix.sum(),3)

accuracy
````

````python
# Grafico de matriz de confusion

ax=sns.heatmap(cf_matrix, annot=True, cmap='Blues')
ax.set_title('Confusion Matrix \n Accuracy Score: {}'.format(accuracy))
ax.set_ylabel('Actual Values')
ax.yaxis.set_ticklabels(['False', 'True'])
ax.set_xlabel('Predicted Values')
ax.xaxis.set_ticklabels(['False', 'True']);
````

````python
# Cambio en los nombres de los labels/ticks

ax=sns.heatmap(cf_matrix, annot=True, cmap='Blues')
ax.set_title('Confusion Matrix \n Accuracy Score: {}'.format(accuracy))
ax.set_ylabel('Actual Values')
ax.yaxis.set_ticklabels(['0', '1'])
ax.set_xlabel('Predicted Values')
ax.xaxis.set_ticklabels(['0', '1']);
````
